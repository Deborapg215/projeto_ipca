```
import requests

link = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.10843/dados?formato=json'
requisicao = requests.get(link)
informacoes = requisicao.json()
print(informacoes)
```

```
%sql

select
round(avg(`Valor_IPCA_%`), 2) as media_5_anos
from
tabela_10_anos_clean
where
year(data) in (2018, 2019, 2022, 2023, 2024)
```

```
%sql

SELECT
  year(data) AS Ano,
  ROUND(SUM(`Valor_IPCA_%`), 2) AS IPCA_Anos_Pandemia_Acumulado
FROM
  tabela_10_anos_clean
WHERE
  year(data) IN (2020, 2021)
  AND month(data) BETWEEN 1 AND 12
GROUP BY
  year(data)
ORDER BY
  Ano
```

```
%sql

SELECT
  year(data) AS Ano,
  CASE month(data)
    WHEN 1 THEN 'Janeiro'
    WHEN 2 THEN 'Fevereiro'
    WHEN 3 THEN 'Março'
    WHEN 4 THEN 'Abril'
    WHEN 5 THEN 'Maio'
    WHEN 6 THEN 'Junho'
    WHEN 7 THEN 'Julho'
    WHEN 8 THEN 'Agosto'
    WHEN 9 THEN 'Setembro'
    WHEN 10 THEN 'Outubro'
    WHEN 11 THEN 'Novembro'
    WHEN 12 THEN 'Dezembro'
  END AS Mes,
  ROUND(SUM(`Valor_IPCA_%`), 2) AS IPCA_Mensal_Pandemia
FROM
  tabela_10_anos_clean
WHERE
  year(data) IN (2020, 2021)
GROUP BY
  year(data), month(data)
ORDER BY
  Ano, month(data)
```

```
%sql
SELECT
  YEAR(data) AS Ano,
  CASE MONTH(data)
    WHEN 1 THEN 'Janeiro'
    WHEN 2 THEN 'Fevereiro'
    WHEN 3 THEN 'Março'
    WHEN 4 THEN 'Abril'
    WHEN 5 THEN 'Maio'
    WHEN 6 THEN 'Junho'
    WHEN 7 THEN 'Julho'
    WHEN 8 THEN 'Agosto'
    WHEN 9 THEN 'Setembro'
    WHEN 10 THEN 'Outubro'
    WHEN 11 THEN 'Novembro'
    WHEN 12 THEN 'Dezembro'
  END AS Mes,
  ROUND(SUM(`Valor_IPCA_%`), 2) AS IPCA_Acumulado_Mensal
FROM
  tabela_10_anos_clean
WHERE
  YEAR(data) >= YEAR(CURRENT_DATE) - 10
  AND MONTH(data) BETWEEN 1 AND 12
GROUP BY
  YEAR(data),
  MONTH(data)
ORDER BY
  Ano,
  MONTH(data)
```
